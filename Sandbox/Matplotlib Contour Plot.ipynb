```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)
X, Y = np.meshgrid(x, y)

Z = np.sin(X) + np.cos(Y)
print(type(Z), Z.shape)
#plt.ion()
plt.figure()
plt.contour(X, Y, Z)
plt.colorbar()
```

```python
plt.contourf(X, Y, Z)
plt.colorbar()
```

```python
plt.contourf(X, Y, Z, levels=24)
```

```python
Z = (X**2 + Y**2 -1)**3 - X**2 + Y**3
plt.contour(X, Y, Z, levels=10)
```
