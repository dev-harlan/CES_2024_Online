```python
# 1
import math
import numpy as np
from matplotlib import pyplot as plt

#1.1

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
xx, yy = np.meshgrid(x, y)

```

```python
#1.2
#print(f"xx type: {type(xx)}, xx shape: {xx.shape}")
plt.gca().set_aspect('equal')
plt.pcolormesh(xx)
```

```python
plt.gca().set_aspect('equal')
plt.pcolormesh(yy)
```

```python
#1.3
r_grid = np.zeros((len(x), len(y)))
phi_grid = np.zeros((len(x), len(y)))

for i in range(len(x)):
    for j in range(len(y)):
        r_grid[i][j] = math.sqrt(x[i]**2 + y[j]**2)
        phi_grid[i][j] = np.arctan2(y[i], x[j])


#1.4
plt.gca().set_aspect('equal')
plt.pcolormesh(r_grid)
plt.show()
plt.gca().set_aspect('equal')
plt.pcolormesh(phi_grid)
plt.show()
```

```python
#1.5
f = np.zeros((len(x), len(y)))

for i in range(len(x)):
    for j in range(len(y)):
        f[i][j] = math.cos(4 * r_grid[i][j])**2 + math.sin(4 * phi_grid[i][j])**2  

plt.gca().set_aspect('equal')
plt.pcolormesh(f)
plt.show()
```

```python
#1.6
f_xmean = f.mean(axis=1)

plt.plot(range(100), f_xmean)
plt.title("Mean of f, x axis")
plt.show()
```

```python
#1.7
f_ymean = f.mean(axis=0)

plt.plot(range(100), f_ymean)
plt.title("Mean of f, y axis")
plt.show()
```

```python
# 2.1 Argo Data
T = np.load('./float_data/T.npy')
S = np.load('./float_data/S.npy')
P = np.load('./float_data/P.npy')
date = np.load('./float_data/date.npy')
lat = np.load('./float_data/lat.npy')
lon = np.load('./float_data/lon.npy')
level = np.load('./float_data/levels.npy')


#2.2
print(f'T: {T.shape}, type {type(T)}')
print(f'S: {S.shape}')
print(f'P: {P.shape}')
print(f'date: {date.shape}')
print(f'lat: {lat.shape}')
print(f'lon: {lon.shape}')
print(f'level: {level.shape}')
```

<b>Observations:</b> <BR>
The Date, Lat and Long are one dimensional arrays with 75 entries, one for each float. 
Level has 78 entries, representing depth measurements. 
The T, S, and P arrays are two dimensional, 78 x 75, indicating that they are indexed to the 1D 
parameters. Specificall;y, 75 entries, one per each float, by 78 measurements from depth level.
