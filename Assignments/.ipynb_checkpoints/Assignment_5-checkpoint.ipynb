```python
import numpy as np
from matplotlib import pyplot as plt

#1
blackRock_Temp= np.genfromtxt('/data9/UN3400/data/brf_temp.txt', delimiter='\t')
test_data
#plt.plot(test_data[:,0], test_data[:,1] )
```

```python
#2-3

N, i = np.shape(blackRock_Temp) #get the dimensions
t = blackRock_Temp[:, 0:1]  # 0:1 in the column keeps 2D array structure
d_obs = blackRock_Temp[:, 1:2] #observed
P = 365.25

# model
X = np.column_stack([
    np.cos(2 * np.pi * t / P),  # First term: cos(2*pi*t/P)
    np.sin(2 * np.pi * t / P),          # Second term: (2*pi*t/P)
    np.ones_like(t)             # Third term: Constant (C)
])

# Solve for the parameters (A, B, C) using least squares
params, _, _, _ = np.linalg.lstsq(X, d_obs, rcond=None)
A, B, C = params

# Print the fitted parameters
print(f'Fitted parameters: A = {A}, B = {B}, C = {C}')
```

```python
d_pre = A * np.cos(2 * np.pi * t / P) + B * np.sin(2 * np.pi * t / P) + C

plt.axis([np.min(t), np.max(t), -25, 35])
plt.plot(t, d_obs, 'o', color="tab:grey", markersize=1)
plt.plot(t, d_pre, '-', color="tab:purple",)
plt.xlabel('t')
plt.ylabel('d')
```
