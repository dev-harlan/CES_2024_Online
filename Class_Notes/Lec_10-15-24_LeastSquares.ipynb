```python
# Least Squares / Curve Fitting

import numpy as np
from matplotlib import pyplot as plt


test_data = np.genfromtxt('/data9/UN3400/data/LS_test_data.txt', delimiter='\t')
```

```python
test_data
```

```python
N, i = np.shape(test_data)
print(np.shape(test_data))
print(N)
t = test_data[:, 0:1]  # 0:1 in the column keeps 2D array structure
dobs = test_data[:, 1:2]
dobs
```

```python
# set up linear equation
M = 2
G = np.zeros((N,M))
G[0:N, 0:1] = np.ones((N, 1))
G[0:N, 1:2] = t
G
```

```python
#mEST = GT*G-1 * Gobs *...?

GTG = np.matmul(G.T, G)
GTd = np.matmul(G.T, dobs) # observation
mest = np.linalg.lstsq(GTG, GTd, rcond=None)[0]  # ax = b ---> x = b/a
mest
```

```python
dpre = np.matmul(G, mest)
e = dobs - dpre
sigmad2 = np.matmul(e.T, e)/(N-M)
Cd = sigmad2 * np.linalg.inv(GTG)
sm = np.sqrt(np.diag(Cd).reshape(-1, 1))
sm
```

```python
print('Estimated solution')
print(np.sqrt(sigmad2))

for k in range(M):
    print(f"model paramter {k}: {mest[k,0]} +/- { 2.0*sm[k, 0]}")
```

```python
fig, ax = plt.subplots()
plt.axis([np.min(t), np.max(t), -4, 4])
plt.plot(t, dpre, 'k-')
plt.plot(t, dobs, 'ro')
plt.xlabel('t')
plt.ylabel('d')
```

```python
fig, ax = plt.subplots()
plt.axis([np.min(t), np.max(t), -0.4, 0.4])
plt.plot(t, e, 'ko-')
plt.xlabel('x')
plt.ylabel('e')
plt.title('Prediction Error')
```

```python
#do it in less lines w/ SCIPY

from scipy import optimize

def linfunc(x, a, b):
    y = a*x + b
    return y

mest_easy = optimize.curve_fit(linfunc, xdata = test_data[:, 0], ydata=test_data[:, 1])[0]
print(mest_easy)
```

```python
# also do it with stats
import statsmodels.api as smod
result = smod.OLS(dobs, G).fit()
print(result.summary())
```
