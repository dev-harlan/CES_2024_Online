```python
data = {'mass':[0.3e24, 4.87e24, 5.97e24],
       'diameter':[4879e3, 12104e3, 12756e3],
        'roation_period':[1407.6, np.nan, 23.9]
       }
data
```

```python
# calc density from volume and add to dataframe
volume = 4/3 * np.pi * (df.diameter/2)**3
volume
```

```python
density = df.mass/volume
density
```

```python
df['volume'] = volume
df['density'] = density
df
```

```python
# add new series w/ different dimensions 
temperature = pd.Series([167, 464, 15, -65], 
                        index=['Mercury', 'Venus', 'Earth', 'Mars'],
                       name = 'temperature'
                       )
temperature
```

```python
# merge series

df.join(temperature)
```

```python
df.join(temperature, how='right') # left col, right col
```

```python
#EARTHQUAKE DATA

earthq_df = pd.read_csv('https://www.ldeo.columbia.edu/~rpa/usgs_earthquakes_2014.csv', 
                  index_col = 'id',
                  parse_dates=['time']
                  )
```

```python
earthq_df
```

```python
earthq_df['country'] = earthq_df.place.str.split(', ').str[-1]
```

```python
earthq_df
```

```python
gb = earthq_df.groupby(earthq_df.country)
len(gb)
```

```python
gb.size()
```

```python
gb.count().tail(20)
```

```python
gb.get_group('Mexico')
```

```python
quakes_small = earthq_df[earthq_df.mag<4]
quakes_large = earthq_df[earthq_df.mag>=4]
quakes_large
```

```python
quakes_large.groupby('country').mag.count().nlargest(20).plot(kind='bar')
```

```python
gb = quakes_large.groupby('country')
gb.size()
```

```python
#list(gb.groups.keys())
display(gb.get_group('Chile').count())
gb.get_group('Chile').head(10)
```
