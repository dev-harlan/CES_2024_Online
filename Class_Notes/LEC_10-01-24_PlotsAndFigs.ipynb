```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.cos(x)
z = np.sin(6*x)

fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(8,4))
axes[0].plot(x,y, linestyle='dashed')
axes[1].plot(x,z, linestyle='dotted')
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(15,3))
ax0, ax1, ax3 = axes
ax0.plot(x,y, linestyle='--')
ax1.plot(x,z, linestyle='dotted', color="m") # r, g, b, c, m, y, k
ax3.plot(y,z, linestyle='dashdot', linewidth='0.75', color="red", marker='o')
ax0.set_xlabel('x')
ax0.set_ylabel('cos(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('6 * sin(x)')
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, x, z, y, z, marker='^')
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(10,3))
ax0, ax1 = axes
ax0.plot(x,y, linestyle='--')
ax1.plot(x,z, linestyle='dotted', color="m") # r, g, b, c, m, y, k

ax0.set_xlabel('x')
ax0.set_ylabel('cos(x)')
ax0.set_xticks(np.array([-np.pi, 0, np.pi]))
ax0.set_xticks(np.arange(-np.pi, np.pi, .3), minor=True)
ax0.set_xticklabels(['$-\pi$', '0', '$\pi$'])
ax0.grid(which='major', linewidth=0.5)


ax1.set_xlabel('x')
ax1.set_ylabel('6 * sin(x)')
```

```python
t = np.arange(0, 10, 0.1)
u = np.exp(t)
v = np.sin(0.2) * np.pi * t
v
```

```python
fig, ax = plt.subplots()
ax.plot(t, u, color='tab:red')
ax.tick_params(axis='y', labelcolor='tab:red')
ax.set_ylabel('Left Axis', color='tab:red')



ax2 = ax.twinx()
ax2.plot(t, v, color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')
ax2.set_ylabel('Right Axis', color='tab:blue')
```

```python
labels = ['first', 'second', 'third']
values = [10, 5, 30]
values2 = [20, 60, 10]

fig, axes = plt.subplots(ncols=2, figsize=(10,5))
axes[0].bar(labels, values)
axes[1].barh(labels, values)
```

```python
fig, ax = plt.subplots()

locs = np.arange(len(labels))
width = 0.2
offset =0.1

ax.bar(locs-offset, values, width, label='values')
ax.bar(locs+offset, values2, width, label='values2')
ax.legend()
ax.set_xticks(locs, labels)
```
