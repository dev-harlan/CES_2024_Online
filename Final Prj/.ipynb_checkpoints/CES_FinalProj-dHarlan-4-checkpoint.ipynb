Background: 
The Greenland Ice Sheet has experienced accelerated ice-mass loss over the last three decades. Ice calving events from marine-terminating glaciers account for roughly 30-50% of all ice mass loss. Glacier calving events in Greenland are detectable through seismic signals, referred to as glacial earthquakes, which make earthquakes an excellent proxy for rates of ice mass loss on the ice sheet.  

Hypothesis:
When plotting glacial earthquakes from a catalog of events from 1993 to present, especially on Greenland's west coast, there appears to be some periodicity in the number of earthquakes recorded on a cycle of 6-9 years. It is possible this period is related to ocean circulation, (NAO or AMOC). To make a preliminary test of this I plan to look at sea surface temperature recorded by NOAA over the same time period and perform some statistical modeling to identify if a correlation can be found.


```python
#packages
import geopandas
import numpy as np
import pandas as pd
import datetime
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import xarray as xr
from scipy import optimize
```

### Accessing and plotting glacial earthquake catalog

```python
quakeData = pd.read_csv('data/GLEA_1993_2023_HC.tsv', sep='\t')
quakeData.head()
```

```python
quakeData['date'] = pd.to_datetime(quakeData[['year', 'month', 'day', 'hour', 'minute', 'second']])
quakeData.set_index('date')
```

```python
# Plot total number of earthquakes per year

# Get groups for western Greenland:
grp_YearCoast = quakeData.groupby(['year', 'region']).size().unstack(fill_value=0)

west_grp = grp_YearCoast['WESTERNGREENLAND']
north_grp = grp_YearCoast['BAFFINBAY']
northwest_grp = pd.concat([west_grp, north_grp]).groupby('year').sum()
#print(northwest_grp)

# The x axis to plot
years = grp_YearCoast.index
t = np.arange(len(years))

# Do plot
fig, ax = plt.subplots(figsize=(10,5))

bar_width = 1/3
ax.bar(t + bar_width, northwest_grp, bar_width,  color='grey')

ax.set_xlabel('Year')
ax.set_ylabel('Total Glacial Earthquakes')
ax.set_title('Glacial Earthquake Events Annually, Western Greenland')
ax.set_xticks(t)
ax.set_xticklabels(years, rotation=90, ha='right')
plt.show
```

### Statistical Modeling

```python

# Convert series from plot to dataframe:
eventsDf = pd.DataFrame({'date': northwest_grp.index, 'count': northwest_grp.values})

# Linear Regression w/ Numpy
t = np.arange(len(years))
slope, intercept = np.polyfit(t, eventsDf.iloc[:, 1], deg=1)
resultLin = slope * t + intercept
print(f"Linear model: Slope = {slope}; Intercept = {intercept}")

# Exponential model
def expfunc(x, a, b, c):
    y = a + b*x**c
    return y

# Get the model parameters
mest_exp = optimize.curve_fit(expfunc, xdata = t, ydata=eventsDf.iloc[:, 1])[0]
print(f"Exponential model parameters: a: { mest_exp[0] }, b: { mest_exp[1] }, c: { mest_exp[2] }")

# Calc the time series using new model parameters
resultExp = expfunc(t, mest_exp[0], mest_exp[1], mest_exp[2])

# sin model
def sinfunc(x, a, b, c):
    # base decadal frequency = 0.1 # f = p/t = 1 period / 10 years = 1/10 = 0.1
    # a = amplitude coefficient
    # b = frequency change as function of t
    # c = wave phase
    y = a*x**0.5 * np.sin(2*np.pi*x*(0.1-x*b) + c)
    return y

# Get the model parameters
mest_sin = optimize.curve_fit(sinfunc, xdata = t, ydata=eventsDf.iloc[:, 1])[0]
print(f"Sinusoidal model parameters: a: { mest_sin[0] }, b: { mest_sin[1] }, c: { mest_sin[2] }")

# Calc the time series using new model parameters
resultSin = sinfunc(t, mest_sin[0], mest_sin[1], mest_sin[2])
```

```python
# Plots

fig, ax = plt.subplots(figsize=(10,5))
plt.plot(t, resultLin, color="tab:blue", label="Linear model")
plt.plot(t, resultExp, color="tab:red", label="Exponential model")
ax.bar(t + bar_width, northwest_grp, bar_width,  color='grey')
ax.set_xlabel('Year')
ax.set_ylabel('Total Glacial Earthquakes')
ax.set_title('Linear vs. Exponential model')
ax.set_xticks(t)
ax.set_xticklabels(years, rotation=90, ha='right')
plt.legend()
plt.show

fig, ax = plt.subplots(figsize=(10,5))
plt.plot(t, resultSin, color="tab:red", label="sine model")
ax.bar(t + bar_width, northwest_grp, bar_width,  color='grey')
ax.set_xlabel('Year')
ax.set_ylabel('Total Glacial Earthquakes')
ax.set_title('Sinusoidal model')
ax.set_xticks(t)
ax.set_xticklabels(years, rotation=90, ha='right')
plt.show
```

```python
# Plot
fig, ax = plt.subplots(figsize=(10,5))
plt.plot(t, resultLin+resultSin, color='tab:red', label="model (linear + sine)")
#plt.plot(x, northwest_grp.values, 'bo', label="observed data")
ax.bar(t + bar_width, northwest_grp, bar_width,  color='grey')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Earthquake Events')
ax.set_title('Composite Model')
ax.set_xticks(t)
ax.set_xticklabels(years, rotation=90, ha='right')
ax.legend()
plt.show
```

### Sea Surface Temperature modeling

```python
# NOAA Monthly mean sea surface data:
# Optimum Interpolation (OI) SST Version 2 product
# https://psl.noaa.gov/data/gridded/data.noaa.oisst.v2.html

# local data
file_path = "data/sst.mnmean.nc"
dataset = xr.open_dataset(file_path)

# sel sea surface temp variable from xarray
sst = dataset['sst']

# Select the nearest grid point to the specified coordinates
lat = 69.0  # off coast of Jakobshavn Glacier
lon = -52.0 + 360  # Convert W longitude to 0-360 scale
sst_west = sst.sel(lat=lat, lon=lon, method='nearest')  # Use the nearest flag

# time range
start_date = '1993-01-01'
end_date = '2023-01-01'
sst_west = sst_west.sel(time=slice(start_date, end_date))

# Convert to NumPy
sst_mmean_values = sst_west.values  # Extract SST values - Remembers each alue is a monthly mean!
time_index = pd.to_datetime(sst_west['time'].values)  # Convert time to pandas datetime

plt.figure(figsize=(12, 6))
plt.plot(time_index, sst_mmean_values, label='JakobShavn Coastal SST', color='blue')
plt.xlabel('Year')
plt.ylabel('Sea Surface Temperature (°C)')
plt.title('Time Series: Sea Surface Temperature near JakobShavn Glacier (1993-2023)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Get the maximum for each year

sst_westYearsGrp = sst_west.groupby("time.year")

maxVals = []
years = list(range(1993, 2023))

for year in years:
    ymax = sst_westYearsGrp[year].values.max()
    maxVals.append(ymax)

#print(maxVals)

# Linear Regression w/ Numpy
sst_t = np.arange(len(years))
slope, intercept = np.polyfit(sst_t, maxVals, deg=1)
sst_resultLinear = slope * sst_t + intercept
print(f"Linear model: Slope = {slope}; Intercept = {intercept}")

# sin model
def sinfunc(x, a, b, c):
    # base decadal frequency = 0.1 # f = p/t = 1 period / 10 years = 1/10 = 0.1
    # a = amplitude coefficient
    # b = frequency change as function of t
    # c = wave phase
    y = a*x**0.5 * np.sin(2*np.pi*x*(0.1-x*b) + c)
    return y

# Get the model parameters
mest_sin = optimize.curve_fit(sinfunc, xdata = sst_t, ydata=maxVals)[0]
print(f"Sinusoidal model parameters: a = { mest_sin[0] }, b = { mest_sin[1] }, c = { mest_sin[2] }")

# Calc the time series using new model parameters
sst_resultSin = sinfunc(sst_t, mest_sin[0], mest_sin[1], mest_sin[2])


# Plot
fig, ax = plt.subplots(figsize=(10,5))

bar_width = 1/3
ax.bar(sst_t + bar_width, maxVals, bar_width, color='gray')
plt.plot(sst_t, sst_resultLinear, color='tab:red', label="Linear model")
plt.plot(sst_t, sst_resultSin, color='tab:purple', label="Sine model")
plt.xlabel('Year')
plt.ylabel('Sea Surface Temperature (°C)')
plt.title('Maximum Annual Sea Surface Temperature, JakobShavn Glacier (1993-2023)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot
fig, ax = plt.subplots(figsize=(10,5))

bar_width = 1/3
ax.bar(sst_t + bar_width, maxVals, bar_width, color='gray')
plt.plot(sst_t, sst_resultLinear + sst_resultSin, color='tab:red', label="composite model")
plt.xlabel('Year')
plt.ylabel('Sea Surface Temperature (°C)')
plt.title('Maximum Annual Sea Surface Temperature, JakobShavn Glacier (1993-2023)')
plt.legend()
plt.grid(True)
plt.show()
```

### Comparison

```python
#normalized
min = (resultLin+resultSin).min()
max = (resultLin+resultSin).max()
geq_normal = (resultLin+resultSin) / (max-min)

min = (sst_resultLinear+sst_resultSin).min()
max = (sst_resultLinear+sst_resultSin).max()
sst_normal = (sst_resultLinear+sst_resultSin) / (max-min)
```

```python

# Plot
fig, ax = plt.subplots(figsize=(10,5))

plt.plot(t, geq_normal, color='tab:red', label="glacial earthquake composite model")
plt.plot(sst_t, sst_normal, color='tab:purple', label="seas surface composite model")

plt.xlabel('Year')
plt.ylabel('vals')
plt.title('Comp')
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusion

The compsite model combines a linear regressions and a sine function to model a rate of increase of some value being modulated by interannual periodicity. 

While there appears to be a close inverse relationship between the normalized models, the models do not quite correlate. The sine model parameter "b" is meant to model a changing wave period, however the glacial earthquake b value is below 1 (a decrease in frequency) while the sea surface temperature frequency appears to be stable (almost exactly 1). Additionally the slopes caculated from linear regression are quite different.

A convincing correlation may require further work, or may not be provable. Additional research has suggest that sea suface temperature may not be the best dataset as subsurface temperatures (which are warmer) be are a more likely driver of ice-mass loss at outlet glaciers.
